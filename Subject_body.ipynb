```python
import pandas as pd
```

```python
df = pd.read_csv("/email_dataset.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
# Split the content column into subject and body columns
df[['Subject', 'Body']] = df['text'].str.split('\n', 1, expand=True)
```

```python
# Drop the original content column
df.drop('text', axis=1, inplace=True)
df.drop('label_num', axis=1, inplace=True)
```

```python
# Save the new DataFrame to a CSV file
df.to_csv('emails_new.csv', index=False)
```
